```python
import pandas as pd
import numpy as np
import math
from itertools import combinations

# Assuming 'solvent_dataset' is a DataFrame containing the solvent data
solvent_dataset = pd.read_excel("C:\\Users\\royka\\OneDrive\\Desktop\\double blends solvent data.xlsx")
solvent_table = pd.DataFrame(solvent_dataset)

# Print the table for all possible two-solvent combinations
all_combinations = list(combinations(solvent_table['Name of Solvents'], 2))
all_combinations_table = pd.DataFrame(all_combinations, columns=['Solvent 1', 'Solvent 2'])
print("\nTable for all possible two-solvent combinations:")
print(all_combinations_table)

# Create a dictionary to store solvent data
solvents = {}
for index, row in solvent_table.iterrows():
    solvent_name = row['Name of Solvents']
    parameters = {
        'dD': row['dD'],
        'dP': row['dP'],
        'dH': row['dH'],
        'Ra(distance)': row['Ra(distance)']
    }
    solvents[solvent_name] = parameters

# Function to calculate Ra distance
def calculate_distance(mixed_parameters):
    dD = mixed_parameters['dD']
    dP = mixed_parameters['dP']
    dH = mixed_parameters['dH']
    distance = math.sqrt(4 * (dD - 17.07) ** 2 + (dP - 4.12) ** 2 + (dH - 3.16) ** 2)
    return distance

# Initialize overall minimum Ra value and corresponding composition
min_ra_overall = float('inf')
min_composition_overall = None

# Create a list to store results
results_data = []

# Loop to find and print results for each combination
for i, selected_combination in enumerate(all_combinations):
    selected_solvent_names = list(selected_combination)
    selected_solvent_data = solvent_table[solvent_table['Name of Solvents'].isin(selected_solvent_names)]

    print(f"\nResults for Combination {i + 1} - Solvents: {selected_solvent_names}")

    # Create a dictionary to store solvent data for the selected combination
    solvents_combination = {}
    for index, row in selected_solvent_data.iterrows():
        solvent_name = row['Name of Solvents']
        parameters = {
            'dD': row['dD'],
            'dP': row['dP'],
            'dH': row['dH'],
            'Ra(distance)': row['Ra(distance)']
        }
        solvents_combination[solvent_name] = parameters

    # Initialize minimum Ra value and corresponding composition for the current combination
    min_ra_table = float('inf')
    min_composition_table = None

    # Loop to find the minimum Ra value and corresponding composition for the current combination
    for composition1 in range(10, 91, 10):
        composition2 = 100 - composition1

        # Retrieve parameter values for each solvent
        solvent_1, solvent_2 = selected_solvent_names
        parameters_1 = solvents_combination[solvent_1]
        parameters_2 = solvents_combination[solvent_2]

        # Calculate the mixed parameters
        mixed_parameters = {
            'dD': parameters_1['dD'] * composition1 / 100 + parameters_2['dD'] * composition2 / 100,
            'dP': parameters_1['dP'] * composition1 / 100 + parameters_2['dP'] * composition2 / 100,
            'dH': parameters_1['dH'] * composition1 / 100 + parameters_2['dH'] * composition2 / 100,
            'Ra(distance)': parameters_1['Ra(distance)'] * composition1 / 100 + parameters_2['Ra(distance)'] * composition2 / 100
        }

        ra = calculate_distance(mixed_parameters)

        # Update minimum Ra and corresponding composition for the current combination
        if ra < min_ra_table:
            min_ra_table = ra
            min_composition_table = (composition1, composition2)

    # Update overall minimum Ra value and corresponding composition
    if min_ra_table < min_ra_overall:
        min_ra_overall = min_ra_table
        min_composition_overall = min_composition_table

    # Append the results for the current combination to the results_data list
    results_data.append({
        'Combination': selected_solvent_names,
        'Overall Minimum Ra': round(min_ra_table, 2),
        'Corresponding Composition': min_composition_table
    })

# Convert results_data list to DataFrame and sort based on 'Overall Minimum Ra'
results_table = pd.DataFrame(results_data).sort_values(by='Overall Minimum Ra')

# Print the sorted results table
print("\nResults for all combinations (Sorted by Overall Minimum Ra):")
print(results_table)
```
